```python
x = dataframe.SL
y = dataframe.PW
```

Scatter plot of SL vs PW

```python
plt.scatter(x,y)
plt.xlabel('SL')
plt.ylabel('PW')
plt.show()
```

Histogram of PW

```python
plt.hist(dataframe['PW'],bins=20,edgecolor ='k')
plt.xlabel('PW')
plt.ylabel('Frequency')
plt.show()
```

Bar plot of iris vs SW

```python
plt.bar(dataframe['Classification'],dataframe['SW'])
plt.ylabel('SW')
plt.xticks(rotation=45)
plt.show()
```
